```python
import pandas as pd
from heap import Heap

try:
    # 1. 파일을 읽을 때 열 이름을 강제로 지정합니다 (이름, 연도, 월, 일 순서)
    df = pd.read_csv('birthday.csv', names=['이름', '연도', '월', '일'], header=None)
    
    h = Heap()
    for _, row in df.iterrows():
        try:
            # 2. 데이터에 숫자가 아닌 불순물이 섞여있을 경우를 대비해 처리
            y = int(row['연도'])
            m = int(row['월'])
            d = int(row['일'])
            date_val = y * 10000 + m * 100 + d
            h.insert((date_val, row['이름']))
        except:
            continue # 숫자로 변환 안 되는 줄은 그냥 넘어감

    # 3. 출력 부분
    print("=== 생일이 느린 친구들 TOP 10 ===")
    if h.isEmpty():
        print("데이터를 찾을 수 없습니다. csv 파일 형식을 확인해주세요.")
    else:
        for i in range(1, 11):
            person = h.deleteMax()
            if person:
                print(f"{i}위: {person[1]} ({person[0]})")
            else:
                break
except Exception as e:
    print(f"오류 발생: {e}")
```

```python
from circularDoublyLinkedList import CircularDoublyLinkedList

L = CircularDoublyLinkedList()

L.append(["이시현", 2006, 1, 25])
L.append(["이예지", 2005, 1, 4])
L.append(["이유민", 2005, 7, 11])
L.append(["이윤서", 2006, 12, 27])
L.append(["이주은", 2006, 7, 5])
L.append(["이채영", 2006, 4, 18])
L.append(["장채은", 2005, 5, 8])
L.append(["정희원", 2006, 12, 21])
L.append(["조은서", 2005, 7, 19])
L.append(["진가연", 2005, 11, 12])
L.append(["최윤지", 2006, 8, 30])

print("## 과제 4: 우리 조 친구들 명단 ##")
print("-" * 30)

for person in L:
    name = person[0]
    try:
        if person[1] and person[2] and person[3]:
            birth = f"{person[1]}년 {person[2]}월 {person[3]}일"
        else:
            birth = "생년월일 정보 불충분"
    except:
        birth = "데이터 오류"
        
    print(f"이름: {name} | 생일: {birth}")
```
