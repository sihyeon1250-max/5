```python
import pandas as pd
from heap import Heap

try:
    # 1. 파일을 읽을 때 열 이름을 강제로 지정합니다 (이름, 연도, 월, 일 순서)
    df = pd.read_csv('birthday.csv', names=['이름', '연도', '월', '일'], header=None)
    
    h = Heap()
    for _, row in df.iterrows():
        try:
            # 2. 데이터에 숫자가 아닌 불순물이 섞여있을 경우를 대비해 처리
            y = int(row['연도'])
            m = int(row['월'])
            d = int(row['일'])
            date_val = y * 10000 + m * 100 + d
            h.insert((date_val, row['이름']))
        except:
            continue # 숫자로 변환 안 되는 줄은 그냥 넘어감

    # 3. 출력 부분
    print("=== 생일이 느린 친구들 TOP 10 ===")
    if h.isEmpty():
        print("데이터를 찾을 수 없습니다. csv 파일 형식을 확인해주세요.")
    else:
        for i in range(1, 11):
            person = h.deleteMax()
            if person:
                print(f"{i}위: {person[1]} ({person[0]})")
            else:
                break
except Exception as e:
    print(f"오류 발생: {e}")
```

## Q2. 배열 $A[0 \dots n-1]$을 대상으로 하는 스며내리기(percolate-down) 알고리즘의 최악 및 최선의 경우 시간 복잡도를 $\Theta$-표기로 나타내시오.

**답변:**

스며내리기 알고리즘은 특정 노드에서 시작하여 자식 노드와 값을 비교하며 힙의 성질을 만족할 때까지 아래로 내려가는 방식입니다.
이 과정은 트리의 높이 $h$ 에 비례하며, $n$개의 노드를 가진 완전 이진 트리의 높이는 $\log_2 n$ 입니다.

### 최악의 경우 (Worst Case): $\Theta(\log n)$

- **상황:** 시작 노드의 값이 자식 노드들보다 현저히 작아(최대 힙 기준), 루트 노드에서 시작하여 트리의 가장 아래 단계인 리프 노드(Leaf Node)까지 끝까지 내려가며 비교와 교환을 반복해야 하는 경우입니다.

### 최선의 경우 (Best Case): $\Theta(1)$

- **상황:** 시작 노드의 값이 이미 자식 노드들보다 커서, 단 한 번의 비교만 수행하고 이동 없이 그 자리에서 즉시 알고리즘이 종료되는 경우입니다.

## Q3. 힙에서 원소 삭제는 루트 노드만 대상으로 하지만, 테스트 목적으로 힙의 맨 마지막 원소를 삭제하는 작업을 요구한다면 간단한 일인가?

**답변:**

네, **매우 간단한 일**입니다.

### 힙의 구조적 특징

힙은 **완전 이진 트리(Complete Binary Tree)** 형태를 유지해야 합니다.
완전 이진 트리에서 노드가 추가되거나 삭제될 때, 구조를 깨뜨리지 않고 삭제할 수 있는 유일한 노드는 배열의 가장 마지막 인덱스에 위치한 **맨 마지막 원소**입니다.

### 힙 성질 유지

보통 루트 노드를 삭제할 때는 마지막 원소를 루트로 옮긴 후 스며내리기(percolate-down)를 통해 힙 성질을 복구해야 하지만,
맨 마지막 원소 자체를 삭제할 때는 그 노드가 자식 노드를 가질 수 없는 **리프 노드**이므로,
단순히 삭제만 해도 나머지 노드들 사이의 부모-자식 간 크기 관계(Heap Property)가 그대로 유지됩니다.

### 구현의 용이성

배열로 구현된 힙에서 맨 마지막 원소를 삭제하는 것은 배열의 크기를 나타내는 변수(`__numItems` 등)를 1 감소시키거나,
파이썬 리스트의 `pop()` 메서드를 호출하는 것으로 $\Theta(1)$ 시간 내에 간단히 해결됩니다.
